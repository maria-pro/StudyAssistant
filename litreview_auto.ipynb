```python
import requests
import pandas as pd
from dotenv import load_dotenv

from pprint import pprint
import os

load_dotenv()
LENS_API=os.getenv("LENS_API")
GEMINI_API=os.getenv("GEMINI_API")

SCIVAL_API=os.getenv("SCIVAL_API")
SCIVAL_API_INS=os.getenv("SCIVAL_API_INS")
```

Get collection from lens

```python
#lens API ref

def get_collection(collection_id, query=None, size=10,sort=None):
    url=f"https://api.lens.org/collections/{collection_id}"

    params={
        "token":LENS_API,
        "size": size
    }

    if query:
        params["query"] = query

    if sort:
        params["sort"]=sort

    response=requests.get(url=url, params=params)
    data=response.json()
    return data

```

```python
test=get_collection("231151")

```
